```python
# Session 4: Assignment 1
# Task-1: 
# 1. Write a function to find moving average in an array over a window:
# Test it over [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150] and window of 3.

def moving_average(l, N):
    sum = 0
    result = list( 0 for x in l)
 
    for i in range( 0, N ):
        sum = sum + l[i]
        result[i] = sum / (i+1)
 
    for i in range( N, len(l) ):
        sum = sum - l[i-N] + l[i]
        result[i] = sum / N
 
    return result

l=[3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150]
y = moving_average(l, 3)
print(y)
```

```python
# Session 4: Assignment 1
# Task-1: 
# 1. Write a function to find moving average in an array over a window:
# Test it over [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150] and window of 3.

# Pandas method of evaluating 'Moving Average' with window size = 3
import pandas as pd

# Create data
array_data = {'data': [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150]}

# Create dataframe
df = pd.DataFrame(array_data)

# View dataframe
print(df)

# Step-1: Now I am calculating the moving average by taking the first 3 values and averaging them, 
# Step-2: then drop the first and add the third, etc.
df.rolling(window=3).mean()
```

```python
# Session 4: Assignment 1
# Task-2:
# 1. How-to-count-distance-to-the-previous-zero
# For each value, count the difference back to the previous zero (or the start of the Series,
# whichever is closer)
# create a new column 'Y'
# Consider a DataFrame df where there is an integer column 'X'

import pandas as pd
# df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})
# df
# df_series = df['X']
s = pd.Series([7, 2, 0, 3, 4, 2, 5, 0, 3, 4])
bs = s.eq(0) # It will generate a Boolean Series
print(bs)
bscum = bs.cumsum() # It will generate a series of same length 
                    # but with the cumulative sum
print(bscum)

bscum_mask = bscum.mask(s.eq(0))
print(bscum_mask)
bscum_mask_grouped = s.groupby(bscum_mask)
print(bscum_mask_grouped)

bmgc = bscum_mask_grouped.cumcount()
print(bmgc)

bmgc_incr_1 = bmgc+1
print(bmgc_incr_1)

distance_from_prev_zero = bmgc_incr_1.mask(s.eq(0), 0)
print(distance_from_prev_zero)

dist_prev_zero = distance_from_prev_zero.tolist()
print(dist_prev_zero)
```

```python
# Session 4: Assignment 1
# Task-2:
# 2. Create a DatetimeIndex that contains each business day of 2015 and 
# use it to index a Series of random numbers.

import pandas as pd
import numpy as np
from numpy.random import randn as rn

drg = pd.date_range('2015-01-01', '2015-12-31', freq='B')
All_busines_date = [d.date() for d in drg]
print("All Business Dates in 2015: ",All_busines_date)

time_stamp = pd.Timestamp('2015-12-31', 'B')
print("Last Day's Time Stamp: ",time_stamp)

Last_date = time_stamp.date()
print("Last Date: ",Last_date)

nbc = np.busday_count(All_busines_date, Last_date)
print(drg)
print("The Total No. of Business Days: ",len(nbc))

print("\nDebasish DataFrame with DatetimeIndex\n",'-'*37, sep='')
np.random.seed(101)
random_df = pd.DataFrame(data=np.round(rn(len(nbc),1),2), index= drg, columns= ['Random numerics'])
print(random_df)
```

```python
# Session 4: Assignment 1
# Task-2:
# 3. Find the sum of the values in s for every Wednesday

import pandas as pd
import numpy as np
from numpy.random import randn as rn

random_df.index.name = "Date_TimeIndex"
print(random_df)

random_df['dates'] = pd.to_datetime(random_df.index)                    
random_df['dayofweek'] = random_df['dates'].apply(lambda x: x.dayofweek)
print(random_df)

df_date_categ = random_df['dates'].apply(lambda x: 0 if x.dayofweek in [2] else 1)
print(df_date_categ)

rndf_grouped = random_df.groupby(df_date_categ)
print(rndf_grouped)

rgp = rndf_grouped.sum()
print(rgp)
# print(random_df)
print("The sum of the values in series for every Wednesday: ",rgp['Random numerics'][0])
```

```python
# Session 4: Assignment 1
# Task-2:
# 4. Average For each calendar month

import pandas as pd
import numpy as np
from numpy.random import randn as rn

s2 = random_df.groupby([lambda x: x.month])
print(s2)

s3 = s2.mean()
print("\nDataFrame Average for each calendar month\n",'-'*37, sep='')
print(s3['Random numerics'])
```
